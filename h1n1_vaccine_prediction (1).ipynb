```python
for i in df.columns:
  if df[i].isna().sum()>=0:
    if df[i].dtype=='float64' or df[i].dtype=='int64':
      df[i]=df[i].fillna(df[i].mean())
    elif df[i].dtype=='O':
      df[i]=df[i].fillna(df[i].mode()[0])
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sd=StandardScaler()
```

```python
scaled=pd.DataFrame(sd.fit_transform(df),columns=df.columns)
```

```python
scaled
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=40)
principalComponents=pca.fit_transform(scaled)
principledf=pd.DataFrame(principalComponents)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df_train,df_test,Y_train,Y_test=train_test_split(principledf,Y,test_size=0.2,random_state=123)
```

```python
df_train.shape,df_test.shape,Y_train.shape,Y_test.shape
```

```python
model=LogisticRegression()
model.fit(df_train,Y_train)
```

```python
Ypred=model.predict(df_test)
```

```python
model.score(df_test,Y_test)
```

```python
from sklearn.metrics import  accuracy_score,confusion_matrix,classification_report
```

```python
confusion_matrix(Y_test,Ypred)
```

```python
l=classification_report(Y_test,Ypred)
```

```python
accuracy_score(Y_test,Ypred)
```
